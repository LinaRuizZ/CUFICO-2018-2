```
import ROOT
```

```
from array import array
```

```
MyTree = ROOT.TTree( 'tree1', 'tree para un par de dados' )
```

```
x = array( 'f', [ 0 ] )
y = array( 'f', [ 0 ] )
z = array( 'f', [ 0 ] )
```

```
MyTree.Branch( 'x', x, 'x/F' )
MyTree.Branch( 'y', y, 'y/F' )
MyTree.Branch( 'z', z, 'z/F' )
```

```
from random import randint
from random import gauss
from math import floor
```

```
for i in range(10000):
  x[0] = gauss(5,3)
  y[0] = gauss(5,2)
  z[0] =randint(1,10)
  MyTree.Fill()
```

```
c1 = ROOT.TCanvas( 'c1', 'My First Data example', 200, 10, 700, 900 )
c1.SetFillColor( 18 )
MyTree.Draw("x>>h1(10,0,10)")
c1.Update()
from ROOT import gROOT 
gROOT.GetListOfCanvases().Draw()
```

```
MyTree.Draw("y>>h2(10,0,10)")
c1.Update()
gROOT.GetListOfCanvases().Draw()
```

```
MyTree.Draw("z>>h3(10,0,10)")
c1.Update()
gROOT.GetListOfCanvases().Draw()
```

```
MyTree.Draw("x*y>>h4(20,0,50)")
c1.Update()
gROOT.GetListOfCanvases().Draw()
```

```
MyTree.Draw("x*z>>h5(50,0,100)")
c1.Update()
gROOT.GetListOfCanvases().Draw()
```

```
MyTree.Draw("x*y*z>>h6(20,0,250)")
c1.Update()
gROOT.GetListOfCanvases().Draw()
```

```
MyTree.Draw("x>>h7(20,0,10)","z>=4.5 && z>=5.5")
c1.Update()
gROOT.GetListOfCanvases().Draw()
```

```
MyTree.Draw("x:z>>h8(10,0,10,10,0,10)","y>=4.5 && y<=5.5","colz")
c1.Update()
gROOT.GetListOfCanvases().Draw()
```

```
MyTree.Draw("x:y:z>>h9(10,0,10,10,0,10,10,0,10)","","lego")
c1.Update()
gROOT.GetListOfCanvases().Draw()
```
